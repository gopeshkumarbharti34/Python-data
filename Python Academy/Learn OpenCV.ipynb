```python
import numpy as np
def main():
    img1=np.zeros((512,512,3),np.uint8)
    cv2.line(img1, (0, 99), (99, 0),(255, 0, 0), 2)
    cv2.rectangle(img1,(150,90),(200,170),(0,255,0),2)
    cv2.circle(img1,(60,60),50,(0,0,255),-1)
    cv2.ellipse(img1,(160,260),(50,20),0,0,360,(127,127,127),-1)
    
    # Poly gone 
    points=np.array([[80,2],[125,2],[179,50],[125,100],[50,50]],np.int32)
    points=points.reshape((-1,1,2))
    cv2.polylines(img1,[points],True,(0,255,255))
    
    test1='Test Text'
    cv2.putText(img1, test1, (100,100), cv2.FONT_HERSHEY_SIMPLEX,2,(255,255,0))
    
    
    cv2.imshow("Lena",img1)
    cv2.waitKey(0)
    cv2.destroyWindow("Lena")
if __name__=="__main__":
    main()
    
```

# Mini Project: OpenCV BGR Palette with Trackbars

```python
import cv2
import numpy as np
def emptyFunction(self):
    pass


def main():
    image=np.zeros((512,512,3),np.uint8)
    windowName="Open CV BGR color Palette"
    cv2.namedWindow(windowName)
    
    cv2.createTrackbar("B",windowName,0,255,emptyFunction)
    cv2.createTrackbar("G",windowName,0,255,emptyFunction)
    cv2.createTrackbar("R",windowName,0,255,emptyFunction)
    
    while(True):
        cv2.imshow(windowName,image)
        
        if cv2.waitKey(1)==27:  # 27 is code for esc key 
            break
            
        blue=cv2.getTrackbarPos("B",windowName)
        green=cv2.getTrackbarPos("G",windowName)
        red=cv2.getTrackbarPos("R",windowName)
        
        image[:]=[blue,green,red]
        print(blue,green,red)
        
        
    cv2.destroyAllWindows()
    
    
if __name__=="__main__":
    main()
```

# Mini Project: Mouse events and Interactive Drawing 

```python
import cv2
import numpy as np

# Create Black image ad window 
windowName="Drawing"
img=np.zeros((512,512,3),np.uint8)
cv2.namedWindow(windowName)

#Mouse call back function
def draw_circle(event,x,y,flags,param):
    if event==cv2.EVENT_LBUTTONDBLCLK:
        cv2.circle(img,(x,y),40,(0,255,0),-1)  #40 rdius , (0,255,0) Green color -1 is for filled 
    if event==cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),30,(255,0,0),1)  # 1 for non filled 
    if event==cv2.EVENT_RBUTTONDBLCLK:
        cv2.circle(img,(x,y),30,(0,0,255),-1)
        
# bind the callback function to window 
cv2.setMouseCallback(windowName,draw_circle)

def main():
    while(True):
        cv2.imshow(windowName,img)
        if cv2.waitKey(20)==27:
            break
    cv2.destroyAllWindows()
    
if __name__=="__main__":
    main()
        
```
