```python
# B) The file co2.txt also has the percent CO2 emissions per country data. 
# Make a pie graph to represent that data. Save that as Percent.png.
df=pd.read_csv('co2.txt')
#print(df)
slice1=[]
country=[]
arr=np.array(df)
sp=list(arr)
for i in sp:
    slice1.append(i[2])
    country.append(i[0])

plt.pie(slice1,labels=country,startangle=120,shadow=True,explode=(0,0.2,0,0),autopct='%1.1f%%',radius=0.8)
plt.title('CO2 Emission')
plt.legend()
plt.show()
```

```python
# Top 5 Country graph 
df=pd.read_csv('co2.txt')
#print(df)
list1=[]
x=[]
y=[]
arr=np.array(df)
sp=list(arr)
for i in sp:
    list1.append([i[0],i[1]])
list1.sort(key=lambda x: x[1])
list1.reverse()
for i in list1:
    x.append(i[0])
    y.append(i[1])
x1=x[0:5]
y1=y[0:5]
print(x)
print(y)
plt.bar(x1,y1,color='R')
plt.title('Total CO2 emissions per country',fontweight='bold')
plt.xlabel('Countries',fontsize='xx-large')
plt.ylabel('CO2 emissions (kt) in 2014',fontsize='xx-large')
plt.show()
```
