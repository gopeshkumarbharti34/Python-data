####  Name (Text)
####  Surname (Text)
####  Age (Integer)
####  Gender (Text)
####  MovieName (Text)

```python
import sqlite3
db=sqlite3.connect('FB_Data')

cur=db.cursor()           # Get cursor object

# Drop table if it already exist using execute() method.
if cur.execute("DROP TABLE IF EXISTS actors"):
    print('Drop Command Executed Successfully ')

# Create Table
if cur.execute("CREATE TABLE actors(name TEXT, surname TEXT, age INTEGER, gender TEXT, moviename TEXT, rank INTEGER)"):
    print('Table Created in Film Board Database')


db.commit()
cur.close()

db.close()
```

```python
import sqlite3
import time
db=sqlite3.connect('FB_Data')
cur=db.cursor()
def display():
    if cur.execute("SELECT name,surname,age,gender,moviename,rank from actors"):
        print('*************** Film Board Members******************')
    
    recordsall=cur.fetchall()
    print("---------------------------------------------------------------------------------------------")
    for record in recordsall:
        print(' Actor Name: {0}\n Actor Surname: {1}\n Actor Age: {2}\n Actor Gender: {3}\n Actor Movie Name: {4}\n Actor Ranking: {5}\n'.format(record[0],record[1],record[2],record[3],record[4],record[5]))
        print("---------------------------------------------------------------------------------------------")
    option() 
    
def insert():
    name1=input("Enter Actor Name :")
    surname1=input("Enter Actor Surname :")
    age1=int(input("Enter Actor Age :"))
    gender1=input("Enter Actor Gender :")
    moviename1=input("Enter Actor Movie Name :")
    rank1=int(input("Enter Actor rank :"))
    if cur.execute("INSERT INTO actors(name,surname,age,gender,moviename,rank) VALUES(?,?,?,?,?,?)",(name1,surname1,age1,gender1,moviename1,rank1)):
        print('************ Record Inserted Successfully **************')
    option()  
    
def delete():
    dname=input("Enter Actor Name which you want to Delete from Film Board : ")
    if cur.execute("DELETE FROM actors WHERE name=?",(dname,)):
        print("************* Record Deleted Successfully *****************")
    option() 
        
def updateR():
    uname=input("Enter Name of Actor who's rank you want to update")
    urank=input("Enter New rank: ")
    if cur.execute("UPDATE actors SET rank=? WHERE name=?",(urank,uname)):
        print("********** Record Updated Successfully ************* ")        
    option() 
        
def updateM():
    usurname=input("Enter Name of Actor who's rank you want to update")
    umovie=input("Enter New Movie Name: ")
    if cur.execute("UPDATE actors SET moviename=? WHERE surname=?",(umovie,usurname)):
        print("********** Record Updated Successfully ************* ") 

def exit():
    print("*********** Thank You..! Good Bye...! ***************")
    db.commit()
    cur.close()
    db.close()

def option():
    print("************************ CHOOSE YOUR OPTION *********************************************")
    print("1.Display Film Board \n2.Insert Actor in Film Board \n3.Delete Actor from film Board by Name \n4.Update Rank of Film Actor by Name \n5.Update Movie name of Actor by Surname \n6.Take DataBase BackUp \n7.Exit")
    print("***********************************************************************************************************")
    soption=int(input("Enter Your Option : "))
    
    if(soption==1):
        display()
    elif(soption==2):
        insert()
    elif(soption==3):
        delete()
    elif(soption==4):
        updateR()
    elif(soption==5):
        updateM()
    elif(soption==7):
        exit()
    else:
        print("Sorry You have Entered Invalid Option Wait for a while and Enter Option again:")
        time.sleep(5)
        option()

    
option()
```
