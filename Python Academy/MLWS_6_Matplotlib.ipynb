```python
# compute the x and y coordinates for point on a sine curve
x=np.arange(0,3*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.title("Sine and Cos Wave form")

# Plot the point using matplot
plt.plot(x,y_sin)
plt.plot(x,y_cos,"--k")
plt.show()
```

```python
x=[1,2,3,4]
y=[5,7,4,6]

x2=[1,2,3,4]
y2=[10,14,12,13]

plt.plot(x,y,label='Loss Per Month',color='R')
plt.plot(x2,y2,label='Profit per Month',color='B')

plt.xlabel('Months',fontsize='xx-large')
plt.ylabel('Lakhs in Rupees',fontsize='xx-large')
plt.title('XYZ Company \n Loss-Profit',fontweight='bold')
plt.legend()
plt.show()
```

# BarChart

```python
x=[1,3,5,7]
y=[5,7,4,6]

x2=[2,4,6,8]
y2=[10,14,12,13]

plt.bar(x,y,label='Loss Per Month',color='R')
plt.bar(x2,y2,label='Profit per Month',color='B')

plt.xlabel('Months',fontsize='xx-large')
plt.ylabel('Lakhs in Rupees',fontsize='xx-large')
plt.title('XYZ Company \n Loss-Profit',fontweight='bold')
plt.legend()
plt.show()
```

# Histogram

# Scatter Plot

```python
x1=[1,2,3,4,5,3,4,2,4,3]
y1=[2,1,3,4,2,5,6,2,1,2]


x2=[9,6,8.9,5.8,7,10,9.7,8,9,6]
y2=[7,9,5,8,9,10.5,8,7.9,8,6]

plt.scatter(x1,y1,label='Bangada Fish',color='k',s=25,marker='o')
plt.scatter(x2,y2,label='Maral Fish',color='b',s=50,marker='v')


plt.xlabel('Weight in grams')
plt.ylabel('Length in cm')
plt.title('XYZ Fish Company')
plt.legend()
plt.show()
```

# SatckPlot

```python
days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','k','b','r'],labels=['Sleeping','Eating','Working','Playing'])



plt.xlabel('Days')
plt.ylabel('Total hrs spend')
plt.title('Stack of Day time Utilization')
plt.legend()
plt.show()
```

# Pie Chart

```python
slices=[7,2,8,7]
activities=['Sleeping','Eating','Working','Playing']
cols=['c','m','r','g']
plt.pie(slices,labels=activities,colors=cols,startangle=120,shadow=True,explode=(0,0.2,0,0),autopct='%1.1f%%',radius=0.8)
plt.title('Monday Time Utilization')
plt.legend()
plt.show()
```
