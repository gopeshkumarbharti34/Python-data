# Concatenation

```python
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
                 'B':['B0','B1','B2','B3'],
                 'C':['C0','D1','C2','C3'],
                 'D':['D0','D1','D2','D3']},index=[1,2,3,4])
print(df1)
```

```python
df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
                 'B':['B4','B5','B6','B7'],
                 'C':['C4','D5','C6','C7'],
                 'D':['D4','D5','D6','D7']},index=[1,2,3,4])
print(df2)
```

```python
df3=pd.DataFrame({'A':['A8','A9','A10','A11'],
                 'B':['B8','B9','B10','B11'],
                 'C':['C8','D9','C10','C11'],
                 'D':['D8','D9','D10','D11']},index=[1,2,3,4])
print(df3)
```

```python
print(pd.concat([df1,df2,df3],axis=0))#axis=0 by default 
print('\n')
```

```python
print(pd.concat([df1,df2,df3],axis=0,ignore_index=True))#axis=0 by default 
print('\n')
```

```python
print(pd.concat([df1,df2,df3],axis=1))#axis=0 by default 
print('\n')
```

```python
df4=pd.DataFrame({'A':['A12','A13','A14','A15'],
                 'B':['B12','B13','B14','B15'],
                 'C':['C12','D13','C14','C15'],
                 'D':['D12','D13','D14','D15']},index=[0,1,2,3])
print(df4)
```

```python
print(pd.concat([df1,df4],axis=1))#axis=0 by default 
print('\n')
```

# Merging 

##### ***************************************SQL Joins**********************************************************************
##### Inner Joins :  Both table fetch with only matched values. if any value not matched then it not get fetch 
##### left Joins : Left Table Completely Fetch but From Right Table only matched values fetch 
##### Right Joins : Right Table Completely Fetch but From Left Table only matched values fetch 
##### Outer Joins : Both table fetch if value matched or not, in python blank values filled with NaN
##### SELECT Emp.Name FROM Emp OUTER JOIN Dept  ON Emp.deptid=Dept.deptid 

```python
df_1=pd.DataFrame({'key':['K0','K1','K2','K3','K4'],
                  'A':['A0','A1','A2','A3','A4'],
                  'B':['B0','B1','B2','B3','B4']})

df_r=pd.DataFrame({'key':['K0','K1','K2','K3','K5'],
                  'C':['C0','C1','C2','C3','C5'],
                  'D':['D0','D1','D2','D3','D5']})

print(df_1)
print(df_r)
```

```python
pd.merge(df_1,df_r,how='inner',on='key')
```

```python
pd.merge(df_1,df_r,how='left',on='key')
```

```python
pd.merge(df_1,df_r,how='right',on='key')
```

```python
pd.merge(df_1,df_r,how='outer',on='key')
```

```python
help(pd.merge)
```

```python
df_l=pd.DataFrame({'key1':['K0','K0','K1','K2'],
                   'key2':['K0','K1','K0','K1'],
                  'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3']})

df_r=pd.DataFrame({'key1':['K0','K1','K1','K2'],
                   'key2':['K0','K0','K0','K0'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']})

print(df_l)
print(df_r)

```

```python
pd.merge(df_l,df_r,on=['key1','key2'])  # Default join in inner 
```

```python
pd.merge(df_l,df_r,on=['key1','key2'],how='outer')
```

```python
pd.merge(df_l,df_r,on=['key1'],how='outer')
```

```python
pd.merge(df_l,df_r,on=['key1'],how='outer',suffixes=['_df_l','_df_r'])
```

```python
pd.merge(df_l,df_r,on=['key1'],how='outer',suffixes=('_df_l','_df_r'))
```
