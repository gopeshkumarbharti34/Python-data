```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.net()
```

```python
import pandas as pd
```

```python
tips=pd.read_csv('tips.csv')
```

```python
tips.head(20)
```

```python
help(sns.set())
```

# Displot 

```python
# Distribution of a univariate set of observations
sns.distplot(tips['tip'],kde=1)# to omit the blue curve use kde=False
#use bins=50,40, etc kernel density estimator 
#for i in range(0,9,0.25):
 #   bin
#sns.distplot(tips['tip'],kde=1,bins=[0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6,6.5,7,7.5,8,8.5])
```

# joint Plot 

```python
#jointplot() allows you 
#match up two distplot for bivariate data 
#kind="scatter" / "reg" / "reside"  /" kde" / "hex" options
```

```python
sns.jointplot(x='totbill',y='tip',data=tips,kind="scatter")
```

```python
sns.jointplot(x='totbill',y='size',data=tips,kind="reg")
```

```python
sns.jointplot(x='totbill',y='size',data=tips,kind="hex")
```

# PairPlot

```python
# Pairplot will not pairwise relationships across an entire dataframe 
#(for the numerical columns) and support a color hue 
# argument(for categorial columns)
```

```python
sns.pairplot(tips,hue="sex") # hue='smoker'/'sex'/'tips'
```

```python
sns.pairplot(tips[["totbill","sex"]],hue="sex")
#if you want histogram with color from a category
```

```python
sns.pairplot(tips.iloc[:,[0,3]],hue="smoker")#this time using index 
#to select columns
```

# Correlation 

```python
tips.corr() # to find co-relation in an dataframe
```

```python
tips.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'),axis=1)
#to get it as plot
```

```python
sns.heatmap(tips.corr(),cmap='coolwarm',annot=True)
```

# Barplot 

```python
#These very similar plots allow you to get aggregate data off a categorical
#feature in your data. barplot is a general plot that allows you to aggregate the categorial data based off some function 
# by default the mean 
sns.barplot(x='sex',y='totbill',data=tips)
```

```python
help(sns.barplot)
```

```python
import numpy as np
sns.barplot(x='sex',y='totbill',data=tips,estimator=np.std)
```

# CountPlot

```python
# Some as barplot except the estimator is explicitly counting the 
# number of occurrences. which is why we only pass the x value 
sns.countplot(x='sex',data=tips)
```

```python
help(sns.countplot)
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x='smoker',data=tips)
```

# Boxplot

```python
#are used to show the distribution of categorial data.
# A box plot (ofr box-and-whisker plot) shows  the distribution 
# of quantitative data in a way that facilitates camparision 
#between variables or across levels of a categorial variav=ble 

sns.boxplot(x='day',y='totbill',data=tips,palette='rainbow')
```

```python
# can do entire dataframe with orient='h'
sns.boxplot(data=tips,palette='rainbow')
```

```python
# do smokers eat more?
sns.boxplot(x='smoker',y='totbill',hue='day',data=tips,palette='coolwarm')
```

```python
sns.boxplot(x='day',y='totbill',hue='smoker',data=tips,palette='coolwarm')
```
